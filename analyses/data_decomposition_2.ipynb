```python
import os
import pandas as pd

# verificar ano a ano dados faltantes para cada coluna
# imprimir percentual de dados faltantes para cada coluna

treated_folder = "../dados_tratados/separado"
for station in os.listdir(treated_folder):
    for year in os.listdir(f"{treated_folder}/{station}"):
        df = pd.read_csv(f"{treated_folder}/{station}/{year}")
        missing_percent = (df.isnull().sum() / len(df) * 100).sort_values(ascending=True)
        missing_percent = missing_percent.map(lambda x: f"{int(x)}%")
        missing_percent = missing_percent[missing_percent != "100%"]
        print(f"{station} - {year}")
        print(missing_percent)
        print("\n")
```

```python
import os
import pandas as pd
from tabulate import tabulate
data = []
treated_folder = "../dados_tratados/separado"
for station in os.listdir(treated_folder):
    for year in os.listdir(f"{treated_folder}/{station}"):
        file_path = f"{treated_folder}/{station}/{year}"
        df = pd.read_csv(file_path, usecols=["PM2.5"])
        missing_percent = (df['PM2.5'].isnull().sum() / len(df) * 100)
        missing_percent = f"{int(missing_percent)}%"
        data.append([station, year, missing_percent])
df_missing = pd.DataFrame(data, columns=["Estação", "Ano", "Percentual de PM2.5 Ausente"])
estacao_cascata = df_missing[df_missing["Estação"] == "Cascata"]
estacao_piratininga = df_missing[df_missing["Estação"] == "Piratininga"]
```

```python
print(tabulate(estacao_cascata, headers='keys', tablefmt='pretty'))
```

```python
print(tabulate(estacao_piratininga, headers='keys', tablefmt='pretty'))
```
